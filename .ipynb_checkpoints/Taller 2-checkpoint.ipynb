```python
#Importamos las librerias a utilizar
import numpy as np
import pandas as pd
from sklearn import neighbors, datasets
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import recall_score
from sklearn.metrics import precision_score
from sklearn.metrics import f1_score
```

## Obtenemos la matriz de correlacion

```python
#Se se separan las variables en un nuevo dataframe eligiendo del original solo las variables numericas.
data_corr = data.select_dtypes(include=[np.number])
```

```python
#Teniendo en cuenta que la columna ID no es de interes en el analisis de la correlación se elimina del dataSet
del data_corr['Id']
data_corr.head()
```

```python
#Se filtra las correlaciones que sean mayores a 50% con respecto a nuestra variable objetivo SalePrice
#Se crea una grafica para una mejor visualizacion de los datos
corr = data_corr.corr()
best_corr = corr.index[abs(corr['SalePrice']>0.5)]
plt.subplots(figsize=(12, 8))
best_corr = data[best_corr].corr()
sns.heatmap(best_corr, annot=True)
plt.show()
```

Como se puede observar la variable que mejor correlacion tiene con **SalePrice** es **OverallQuall** con un 79% seguida de cerca por **GrLivArea** con un 71%, **GarageCars** con un 64% y **GarageArea** con un 62% 

### Creación de variable binaria.

Se realiza respectiva conversión de la variable objetivo por una binaria, teniendo en cuenta las condiciones brindadas en el enunciado de la actividad.

```python
data_clean['Target'] = np.where(data_clean['SalePrice']>214000.000000 , 1, 0)
```

```python
#Se eliminan las columnas de ID y SalePrice del dataset
data_clean = data_clean.drop(["SalePrice","Id"],axis=1)
```

```python
#Tomamos los valores en x ^ y
#Tomamos las valriables que mejor correlacicon tiene con SalePrice
y = data_clean['Target']
X = data_clean.loc[:,['OverallQual','GrLivArea','GarageCars',
                 'GarageArea','TotalBsmtSF','1stFlrSF']]
```

### Decision tree Classifier

```python
decision_tree = tree.DecisionTreeClassifier(criterion='entropy',
                                            min_samples_split=20,
                                            min_samples_leaf=5,
                                            max_depth = 4,
                                            class_weight={1:3.5})
```

```python
tree2 = decision_tree.fit(X, y)
```

```python
train_cm = confusion_matrix(y, tree2.predict(X))
train_cm
```

```python
#Exactitud
accuracy_score(y, tree2.predict(X))
```

```python
#Exhaustividad
recall_score(y, tree2.predict(X))
```

```python
#Precisión
precision_score(y, tree2.predict(X))
```

```python
#Valor-F
f1_score(y, tree2.predict(X))
```

### Curva ROC

```python
roc = tree2.predict_proba(X)
```

```python
from sklearn.metrics import roc_curve, auc
# get false and true positive rates
fpr, tpr, thresholds = roc_curve(y, roc[:,0], pos_label=0)
base_fpr, base_tpr, _ = roc_curve(y, [1 for _ in range(len(y))])
# get area under the curve
roc_auc = auc(fpr, tpr)
# PLOT ROC curve
plt.figure(dpi=150)
plt.plot(fpr, tpr, lw=1, color='green', label=f'AUC = {roc_auc:.3f}')
plt.title('ROC Curve for Decision Tree classifier')
plt.rcParams['font.size'] = 16
    # Plot both curves
plt.plot(base_fpr, base_tpr, 'b', label = 'baseline')
plt.legend();
```

Como se puede observar en la grafica de roc anterior, el clasificador de arboles de decision puede funcionar muy bien para predicir la variable objetivo.

### GradientBoostingClassifier
